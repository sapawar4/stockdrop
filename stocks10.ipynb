```python
import pandas as pd
import numpy as np
stocks = pd.read_csv('stocks10.csv', parse_dates=["timestamp"], index_col='timestamp')

#Calculate 1 day drop
stocks["MSFT_drop"] = stocks["MSFT"].pct_change()
stocks["AAPL_drop"] = stocks["AAPL"].pct_change()
stocks["SLB_drop"]  = stocks["SLB"].pct_change()
stocks["AMZN_drop"] = stocks["AMZN"].pct_change()
stocks["TSLA_drop"] = stocks["TSLA"].pct_change()
stocks["XOM_drop"]  = stocks["XOM"].pct_change()
stocks["WMT_drop"]  = stocks["WMT"].pct_change()
stocks["T_drop"]    = stocks["T"].pct_change()
stocks["V_drop"]    = stocks["V"].pct_change()

#get index for largest drop for each ticker

msft = stocks["MSFT_drop"].idxmin()
aapl = stocks["AAPL_drop"].idxmin()
slb  = stocks["SLB_drop"].idxmin()
amzn = stocks["AMZN_drop"].idxmin()
tsla = stocks["TSLA_drop"].idxmin()
xom  = stocks["XOM_drop"].idxmin()
wmt  = stocks["WMT_drop"].idxmin()
t    = stocks["T_drop"].idxmin()
v    = stocks["V_drop"].idxmin()

#Get index list
index_list = [msft, aapl, slb, amzn, tsla, xom, wmt, t, v]

#Melt and Groupby to get Result

Result = pd.melt(stocks.loc[index_list, 'MSFT_drop':].reset_index(), 
                 id_vars=['timestamp'], 
                 var_name='Ticker', 
                 value_name='Drop').sort_values("Drop").groupby(["Ticker"]).first()  

Result.T
```
